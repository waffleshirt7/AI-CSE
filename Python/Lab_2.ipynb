```python
def cal():
    while True:
        p = int(input("Enter p = "))
        t = float(input("Enter t = "))
        r = float(input("Enter r = "))
        i = p * t * r / 100
        print(i)
        x = input("Press Enter for more cal, or type 'esc' to exit: ").strip().lower()
        if x == 'esc':
            break

cal()
```

```python
def cal(x,y,z):
    a = x*y
    v = a*z
    return a,v

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))

av = cal(l,b,h)

print(av)
```
