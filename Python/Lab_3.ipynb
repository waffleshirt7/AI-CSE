WATER JUG PROBLEM


```python
print("Water Jug Problem")
x = int(input("Enter X:"))
y = int(input("Enter Y:"))
while True:
    rno = int(input("Enter the Rule No:"))
    # fill bucket with 4L water
    if rno==1:
        if x<4: x=4
    # fill bucket with 3L water
    if rno==2:
        if y<3: y=3
    # empty bucket x
    if rno==5:
        if x>0: x=0
    # empty bucket y
    if rno==6:
        if y>0: y=0
    # fill bucket x with y content, remaining left on y
    if rno==7:
        if x+y>=4 and y>0: x, y=4, y-(4-x)
    # fill bucket y with x content, remaining left on x
    if rno==8:
        if x+y>=3 and x>0: x, y=x-(3-y),3
    # fill bucket x with y contents
    if rno==9:
        if x+y<=4 and y>0: x,y=x+y, 0
    # fill bucket x with y contents
    if rno==10:
        if x+y<=3 and x>0: x,y=0, x+y
    print("X = ", x)
    print("Y = ", y)
    if x==2:
        print("The result is a Goal state")
        break
```

BREADTH FIRST SEARCH


```python
graph = {
'A' : ['B','C'],
'B' : ['D', 'E'],
'C' : ['F'],
'D' : [],
'E' : ['F'],
'F' : []
}

visited = [] # List to keep track of visited nodes.
queue = [] #Initialize a queue

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)
    while queue:
        s = queue.pop(0)
        print(s, end = " ")
        for neighbour in graph[s]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)


bfs(visited, graph, 'A')
```

DEPTH FIRST SEARCH


```python
graph = {
'A' : ['B','C'],
'B' : ['D', 'E'],
'C' : ['F'],
'D' : [],
'E' : ['F'],
'F' : []
}

visited = set() # Set to keep track of visited nodes.

def dfs(visited, graph, node):
    if node not in visited:
        print (node,end=" ")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

# Driver Code
dfs(visited, graph, 'A')
```
